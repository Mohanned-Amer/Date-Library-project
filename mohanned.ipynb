```python
import pandas as pd
import matplotlib.pyplot as plt

# بيانات الطلاب
data = {
    "الطالب": ["أحمد", "محمد", "علي", "خالد", "سامي"],
    "الدرجة": [85, 60, 95, 45, 78]
}

df = pd.DataFrame(data)

# تحديد النجاح والرسوب
df["النتيجة"] = df["الدرجة"].apply(lambda x: "ناجح" if x >= 50 else "راسب")

# عرض البيانات
print("بيانات الطلاب:\n")
print(df)

print("\nالإحصائيات:")
print("متوسط الدرجات:", df["الدرجة"].mean())
print("أعلى درجة:", df["الدرجة"].max())
print("أقل درجة:", df["الدرجة"].min())

# إعداد حجم الرسم
plt.figure(figsize=(8,5))

# رسم الأعمدة باللون الأحمر
bars = plt.bar(df["الطالب"], df["الدرجة"], color="red", edgecolor="black")

# خط النجاح
plt.axhline(y=50, color="blue", linestyle="--", linewidth=2, label="حد النجاح")

# كتابة الدرجة فوق كل عمود
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 1,
        f"{height}",
        ha="center",
        fontsize=10,
        fontweight="bold"
    )

# العناوين
plt.title("درجات الطلاب", fontsize=16, fontweight="bold")
plt.xlabel("الطلاب")
plt.ylabel("الدرجة")
plt.ylim(0, 100)

plt.legend()
plt.grid(axis="y", linestyle=":", alpha=0.5)

plt.show()
```
